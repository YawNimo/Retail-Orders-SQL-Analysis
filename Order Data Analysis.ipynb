```python
import pandas as pd
```

```python
df = pd.read_csv("Downloads/orders.csv", na_values = ['Not Available', 'unknown']) # switch these into null values to keep it clean
df.head()
```

```python
df['Ship Mode'].unique()
```

```python
# Renaming the columns with lower case + underscores
df.rename(columns ={'Order Id': 'order_id', 'City':'city'}, inplace = True)
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_') #This replaces the spaces with underscores
df.head(5)
```

```python
#derive a discount column, sale price and profit
df['discount'] = df['list_price']*df['discount_percent']/100
df['sale_price'] = df['list_price']- df['discount']
df['profit'] = df['sale_price']- df['cost_price']
df
```

```python
#converted order date from object data type to datetime
df['order_date']= pd.to_datetime(df['order_date'],format = "%Y-%m-%d")
df.dtypes
```

```python
# dropped these columns as it doesn't serve much in our analysis 
df.drop(columns=['list_price','cost_price','discount_percent'])
df
```

```python
import sqlalchemy as sal

#M
username = 'root'
password = ''  # Leave it blank if no password
host = 'localhost'
port = '3306'
database = 'retail_order_upload'  # Replace this with your actual DB name

# Created an engine
engine = sal.create_engine(f'mysql+pymysql://{username}:{password}@{host}:{port}/{database}')

conn = engine.connect()

# Uploaded my DataFrame to MYSQL
df.to_sql(name='df_orders', con=conn, if_exists='replace', index= False)

print("Data uploaded successfully!")
```
